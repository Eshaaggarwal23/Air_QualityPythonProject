```python
df= pd.read_csv('/content/sample_data/Air_Quality.csv',parse_dates=['last_update'])
```

```python
df.info()
```

```python
df['pollutant_avg']= (df['pollutant_max'] + df['pollutant_min'])/2
```

```python
df.head(6)
```

```python
maxValueRecorded= df[df['pollutant_max']== df['pollutant_max'].max()]
```

```python
maxValueRecorded.head()
```

```python
pollutantID = df.groupby('state')['city','station','pollutant_id'].sum().reset_index()
```

```python
pollutantID.info()
```

```python
#plotting pollutant_avg with ID
Pollutant = df.groupby('ID')['pollutant_avg'].sum().reset_index()
Pollutant.head()
sns.pointplot(x= Pollutant['ID'],y= Pollutant['pollutant_avg'])
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90,fontsize=8)
sns.pointplot(x= Pollutant['ID'],y= Pollutant['pollutant_avg'],color='green')
plt.show()
```

```python
Pollutantclassify= df.groupby('city')['pollutant_avg'].sum().reset_index().head(15)
Pollutantclassify
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=Pollutantclassify['pollutant_avg'],y=Pollutantclassify['city'])
plt.show()
```

```python
PollutantType= df.groupby('pollutant_id')['pollutant_avg'].sum().reset_index().head(15)
PollutantType
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=PollutantType['pollutant_avg'],y=PollutantType['pollutant_id'])
plt.show()
```

```python
Delhi= df[df['state']=='Delhi']
Delhi= Delhi.groupby('station')['pollutant_min','pollutant_max','pollutant_avg'].sum().reset_index()
Delhi.head()
```

```python
Maharashtra= df[df['state']=='Maharashtra']
Maharashtra= Maharashtra.groupby('station')['pollutant_min','pollutant_max','pollutant_avg'].sum().reset_index()
Maharashtra.head()
```

```python
plt.figure(figsize=(8,6))
plt.title("visualisation")
plt.scatter(Delhi.index,Delhi['pollutant_avg'],color='red',label="Delhi",marker='*')
plt.scatter(Maharashtra.index,Maharashtra['pollutant_avg'],color='blue',label="Maharashtra",marker='*')
plt.legend(loc=2)
plt.show()
```

```python
Karnataka= df[df['state']=='Karnataka']
Karnataka= Karnataka.groupby('station')['pollutant_min','pollutant_max','pollutant_avg'].sum().reset_index()
Karnataka.head()
```

```python
Uttar_Pradesh= df[df['state']=='Uttar_Pradesh']
Uttar_Pradesh= Uttar_Pradesh.groupby('station')['pollutant_min','pollutant_max','pollutant_avg'].sum().reset_index()
Uttar_Pradesh.head()
```

```python
plt.figure(figsize=(8,6))
plt.title("visualisation2")
plt.scatter(Delhi.index,Delhi['pollutant_avg'],color='red',label="Delhi")
plt.scatter(Maharashtra.index,Maharashtra['pollutant_avg'],color='blue',label="Maharashtra")
plt.scatter(Karnataka.index,Karnataka['pollutant_avg'],color='green',label="Karnataka")
plt.scatter(Uttar_Pradesh.index,Uttar_Pradesh['pollutant_avg'],color='yellow',label="Uttar_Pradesh")
plt.legend(loc=1)
plt.show()
```
